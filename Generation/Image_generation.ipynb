```python
import os
from datetime import datetime
import torch
from diffusers import StableDiffusionPipeline

```

```python
OUTPUT_DIR = "generated_images"
os.makedirs(OUTPUT_DIR, exist_ok=True)
```

```python

MODEL_ID = "runwayml/stable-diffusion-v1-5"
pipe = StableDiffusionPipeline.from_pretrained(
    MODEL_ID,
    torch_dtype=torch.float32
)

pipe = pipe.to("cpu")

# Memory optimizations for CPU
pipe.enable_attention_slicing()
```

```python
def generate_crochet_image(user_prompt):
    """
    Generates ecommerce-style crochet product image (CPU only)
    """

    final_prompt = f"""
    Create a realistic professional ecommerce product photo of:
    {user_prompt}

    Requirements:
    - handmade crochet product
    - isolated object
    - full product clearly visible
    - centered composition
    - clean white background
    - professional product photography
    - soft studio lighting
    - high detail texture
    - sharp focus
    - no text
    - no watermark
    - square product image
    """

    image = pipe(
        final_prompt,
        num_inference_steps=25,
        guidance_scale=8
    ).images[0]

    filename = f"crochet_{datetime.now().strftime('%Y%m%d_%H%M%S')}.png"
    path = os.path.join(OUTPUT_DIR, filename)

    image.save(path)

    return path
```

```python
# Example usage
if __name__ == "__main__":
    user_prompt = "harry potter plushie"
    print(generate_crochet_image(user_prompt))
```
